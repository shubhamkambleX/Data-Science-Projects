# Preprocessing the data

```python
df.isnull().sum()
```

```python
# In this we will check how much percentage the nan value have for each feature
# 1) step 1 - make the list of features having the nan values
features_nan_val = [features for features in df.columns if df[features].isnull().sum()>1]
# 2)step 2 - Print the missing values and there mean
for feature in features_nan_val:
    print(feature,np.round(df[feature].isnull().mean() *100,2), '% missing values')
```

```python
df['Product_Category_2'] = df['Product_Category_2'].fillna(df['Product_Category_2'].mean())
```

```python
df['Product_Category_3'] = df['Product_Category_3'].fillna(df['Product_Category_3'].mean())
```

```python
df.isnull().sum()
```

```python
df = pd.get_dummies(df,columns=['Age','Gender','City_Category','Marital_Status','Stay_In_Current_City_Years'], drop_first=True) 
```

```python
df.head()
```

```python
df.drop(['User_ID','Product_ID'],axis=1,inplace=True)
```

# input split

```python
X = df.drop(['Purchase'],axis=1)
y = df['Purchase']
```

```python
from sklearn.model_selection import cross_val_score,train_test_split
from sklearn.metrics import mean_squared_error,confusion_matrix,classification_report

def model_train(model,X,y):
#     train test split
    X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.25)
    model.fit(X_train,y_train)
    
#     predict the model
    pred = model.predict(X_test)
    
#     cross validation
    cv_score = cross_val_score(model,X,y,scoring='neg_mean_squared_error',cv=5)
    cv_score = np.abs(np.mean(cv_score))
    
    print("Result")
    print("Mean Square error",np.sqrt(mean_squared_error(y_test,pred)))
    print("CV Score",np.sqrt(cv_score))
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(normalize=True)
```

```python
model_train(model,X,y)
coef = pd.Series(model.coef_,X.columns).sort_values()
coef.plot(kind='bar',title="Model Coefficent")
```
